```python
# what problem makes outliers?
# statistics problems creates for example i find mean or median or mode then outliers effects our result make wrong result
# i am finding average of data
d=[5,6,7,5,6,8,5,4]
mean=sum(d)/len(d)
print(mean)
```

```python
# now i add outlier like 100 i add in data it wil effect our whole result
d1=[5,6,7,5,6,8,5,4,100]
mean=sum(d1)/len(d1)
print(mean)
```

```python
# we can remove outliers using iqr and zscore
# removing outliers using IQR
# IQR stans for inter quantile range 
# we understand quantie with small example if 100 persons are giving exam and my rank is 4 than we have 96 precentile
# and  we have 3 quantile and our 1st quantile is on 25% precentile of data,and third is on 75% precentile of data
# iqr is inter quantile range and it is taken subtracting q3-1,iqr=q3-q1
# minimum range min=Q1-(1.5 * iqr) and data below from min is outlier(we can have outlier in mininmum)
# maximum range max=Q3+(1.5 * iqr) and data above from max is outlier(we can have outlier in maximun)
```

```python
od.shape
```

```python
# now we are detecting outliers
# finding first Quantile
q1=od['CoapplicantIncome'].quantile(0.25)
q1
```

```python
q3=od['CoapplicantIncome'].quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
# finding minimum range of data
min_range=q1-(1.5 * IQR)
min_range
```

```python
max_range=q3+(1.5 * IQR)
max_range
```

```python
# removing outliers from the origina data
od=od[od['CoapplicantIncome'] <= max_range]
```

```python
od.shape#no outlers removes our rows 614 to 596
```

```python
# now see the outliers using boxplot
```

```python
sns.boxplot(x='CoapplicantIncome',data=od)
plt.show()#now our outliers are removed there are 2 or 3 outliers there which is not problem
```

```python
# zcore outliers identification and removing process
# formula of zscor=x-mean/standarddeviation no we find the zscore value for each data of coloumn,and set the range of z score 
# the range is till 3rd stanard deviation we 99 % data on 3rd standard deviation
# we get minimum range mean - 3rd standard deviation,mean + 3rdstandard deviation
# and values after 3rd standard deviation on distribution curve are outliers 
```
